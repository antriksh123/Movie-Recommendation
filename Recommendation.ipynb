```python
import numpy as np
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
credits.head(1)['cast']
```

```python
movies = movies.merge(credits, on = 'title')
# merging on the basics of title
movies.shape
```

```python
movies.info()
```

```python
# genres
# id
# keywords
# title 
# overview
# cast  
# crew

movies = movies[['id', 'title', 'keywords', 'overview',
       'cast', 'crew', 'genres']]
```

```python
movies.info()
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
# make column in correct format
movies.iloc[0].genres
```

```python
# we need to convert this list of dictionary into an array of genres
import ast 
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
# ast.literal_eval function is used to convert string into actual format as in the above case obj is taken as a string by the function
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

```python
def extract_top3(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            L.append(i['name'])
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(extract_top3)
movies.head(1)
```

```python
# now we want to extract director name
def director_job(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(director_job)
movies.head()
```

```python
 # converting movies overview from string to list 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# data sorted 
# merging but first we remove spaces from names
# to make tags
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies.head(1)
```

```python
# merge 
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast']+ movies['overview'] + movies['crew']
movies.head(1)                
```

```python
newdf = movies[['id', 'title', 'tags']]
```

```python
newdf.head()
```

```python
# converting tags into string
newdf['tags'] = newdf['tags'].apply(lambda x:" ".join(x))
newdf.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
newdf['tags'] = newdf['tags'].apply(stem)
newdf['tags'][0]
```

```python
# coverting all into lowercase 
newdf['tags'] = newdf['tags'].apply(lambda x:x.lower())
newdf.head()
```

```python
# we are done with data 
# now we are making model
# converting text to vector
# bag of words technique
# completely based on maths ---> vectorisation 
# we are not going tp consider stop words -- used for sentence formation
# we are going to use sklearn for all these operations

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(newdf['tags']).toarray()
# converting this to numpy array 
vector
```

```python
cv.get_feature_names()
```

```python
# apply steming for words cleaning -- done above 
# calculating angular distance between two vectors

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vector)
```

```python
similar[0]
```

```python
#using enumerate function to convert array into truples
sorted(list(enumerate(similar[0])), reverse = True, key = lambda x:x[1])
```

```python
sorted(list(enumerate(similar[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_idx = newdf[newdf['title'] == movie].index[0] 
    distance = similar[movie_idx]
    movies_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
